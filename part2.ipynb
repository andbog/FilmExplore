# What movies are the best

To answer this we will look into ratings given to films

```python
import numpy as np
import pandas as pd
```

To answer this question mainly the ratings.csv file will be used but I will also use movies.csv to retrieve movie title. First we will calculate some basic statisitics on ratings to choose then the best rated movies

```python
movies = pd.read_csv('movies.csv')
ratings = pd.read_csv('ratings.csv')
```

```python
#ratings_stats=ratings.groupby('movieId').rating.agg(['mean','count',lambda u: np.std(u,ddof=0),'median','min','max'])
ratings_stats.rename(columns={'<lambda>':'std'},inplace=True)
top10=ratings_stats.sort_values(by = 'mean',ascending=False).head(10)
top10
```

Mean seems to be not enough as we are ending up with movies rated by only one user. It's subjective how many users should rate the movie for the rating to make sense, but generally the more the better. I'll choose only movies rated by at least 50 users.

```python
top10=ratings_stats[ratings_stats['count']>50].sort_values(by = ['mean','count'],ascending=[False,False]).head(10)
top10
```

We see now the best rated films but it would be nice to know what films are those 

```python
movies.set_index('movieId',inplace=True)
movies.loc[top10.index,:]
```
