```python
ratings_film=pd.read_pickle('ratings_film.pkl')
```

The 2 function below will allow me to calculate mean and standard deviation (of population) of genre based on composing films

```python
def meanpopul (aggdataframea):
    aggdataframe=aggdataframea.copy()
    aggdataframe['meanW']=aggdataframe['mean']*aggdataframe['count']
    return (aggdataframe['meanW'].sum())/(aggdataframe['count']).sum()

def stdpopul (aggdataframea):
    aggdataframe=aggdataframea.copy()
    aggdataframe['varW']=(aggdataframe['count'])*((aggdataframe['std']**2)+(aggdataframe['mean']**2))
    return math.sqrt((aggdataframe['varW'].sum()/(aggdataframe['count'].sum()))-(meanpopul(aggdataframe)**2))
```

Preparing stats for genres

```python
film_count=[]
rates_count=[]
mean_rating=[]
std_rating=[]
ind=[]

for i in ['Action','Adventure','Animation','Children','Comedy','Crime','Documentary','Drama','Fantasy','Film-Noir','Horror','Musical','Mystery','Romance','Sci-Fi','Thriller','War','Western','NoGenre']:
    temp = ratings_film[ratings_film[i]]
    ind.append(i)
    film_count.append(temp.title.count())
    rates_count.append(temp['count'].sum())
    mean_rating.append(meanpopul(temp))
    std_rating.append(stdpopul(temp))
    
genres_rating=pd.DataFrame({'no of films':film_count,'no of ratings':rates_count,'average rating':mean_rating,'std of rating':std_rating},index=ind)
```

Just adding more indicators: to mark average number of ratings per film for a genre ,upper and lower border for 95% confidence interval (remembering valid ratings are 0-5).

```python
genres_rating['no films vs ratings']=genres_rating['no of ratings']/genres_rating['no of films']
genres_rating['upper'] = genres_rating['average rating'] + norm.ppf(0.975)*genres_rating['std of rating']/(np.sqrt(genres_rating['no of ratings']))
genres_rating.loc[genres_rating['upper']>5,'upper']=5
genres_rating['lower'] = genres_rating['average rating'] + norm.ppf(0.025)*genres_rating['std of rating']/(np.sqrt(genres_rating['no of ratings']))
genres_rating.loc[genres_rating['lower']<0,'lower']=0
genres_rating.sort_values(by='average rating',ascending=False)
```

Let's create some plots to see what do we deal with

```python
genres_rating.loc[:,['average rating','no of films','no of ratings','no films vs ratings']].sort_values(by='average rating').plot.barh(subplots=True,layout=(1,4),figsize=(15,8),legend=None,sharey=True,sharex=False,grid=True)
```

Rating seems not to differ too much among genres (all of them between 3 and 4). I'll check the difference between Romance and Musical. Those seem to be somehow related (there are 344 films that are both Musical and Romance).I will analyse it now only on films that are either on or another. For the comparison simple t-test will be used (as variances might be different Welch's t-test is actually used). Similar test will be conduct on Crime vs Horror

```python
romance=ratings_film.loc[ratings_film['Romance'] & ~ratings_film['Musical'],['mean','count','std']]
musical=ratings_film.loc[ratings_film['Musical'] & ~ratings_film['Romance'],['mean','count','std']]
musical.head()
```

```python
def test_ratings(a,b,level=0.05,twotail=True):
    x1=(meanpopul(a),stdpopul(a),a['mean'].count())
    x2=(meanpopul(b),stdpopul(b),b['mean'].count())
    print("First dataframe has n=%d observations , mean= %f, std= %f" % (x1[2],x1[0],x1[1]))
    print("Second dataframe has n=%d observations , mean= %f, std= %f" % (x2[2],x2[0],x2[1]))
    print("Mean difference of %f" %(abs(x2[0]-x1[0])))
    Teststat=(abs(x2[0]-x1[0]))/(math.sqrt((pow(x1[1],2)/x1[2])+(pow(x2[1],2)/x2[2])))
    degF=min(x1[2],x2[2])-1
    pval=1-t.cdf(Teststat,degF)
    if twotail:
        isDifferent = (pval<(level/2))
    else:
        isDifferent = (pval<level)
        
    print("Value of test statistics equals %f, p-value = %f" %(Teststat,pval))
    
    if isDifferent:
        print("Those 2 are statistically different on significance level of %f" %(level))
    else:
        print("We cannot reject H0 for means being equal at significance level of %f" %(level))
    return isDifferent

test_ratings(musical,romance,level=0.05)
```

```python
horror=ratings_film.loc[ratings_film['Horror'] & ~ratings_film['Crime'],['mean','count','std']]
crime=ratings_film.loc[ratings_film['Crime'] & ~ratings_film['Horror'],['mean','count','std']]
test_ratings(horror,crime)
```
