# Production time vs scoring time

Does the time of the production or the time of scoring influence rating. As ratings may be produced regardless of film. This time ratings.csv would be used as based database. Nontheless, first, production year as well as rating year have to prepare. As we will be working on more than 26M records, we have to stick to as little columns as neccessary to achieve the task. First movies production year. Title in most cases is embeded in title in brackets. However, it's not case for all. Additionally, some dates are unrealistic. As borderders I'm setting 1896 ('Train arrival' of Lumiere brothers) and 2017 (year of last update of this dataset) - even though there may be in the database films that are 'under production', but still those should not yet have been rated.

```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('movies.csv')
movies['year']=pd.to_numeric(movies['title'].str.extract('(\(\d{4}\))',expand=False).str.replace('(','').str.replace(')',''), errors='coerce')
movies.loc[((movies['year']<1896) | (movies['year']>2017)),'year'] = np.nan # 1896 - year of creation of arrival of train, 2018 as the database is from 2018
movies.set_index('movieId',inplace=True)
movies.head()
```

```python
movies['year'].isnull().sum()
```

There are still 228 cases. Let's check what films are those, but having so many films I may leave those one

```python
movies[movies['year'].isnull()].title
```

Just removing unnecessary columns I won't need to free some space for later.

```python
movies.drop(['title','genres'],axis=1,inplace=True)
movies.head()
```

Now, I'm moving to core of the analysis ratings.csv

```python
ratings = pd.read_csv('ratings.csv')
ratings['rate_year']=pd.to_datetime(ratings.timestamp,unit='s').dt.year
ratings.drop(['timestamp'],axis=1,inplace=True)
ratings.head()
```

```python
ratings.set_index(['userId','movieId'],inplace=True)
ratings.head()
```

```python
ratings=ratings.merge(movies,how='left',left_index=True,right_index=True)
ratings.head()
```

```python
ratings['time_diff']=ratings.rate_year-ratings.year
ratings.describe()
```

At this point I'm going to remove points with no production year rating (2953 points) and the ratings done before production (negative time_diff) and correlate the rest against rating.

```python
ratings=ratings[~(ratings['time_diff']<0 | (ratings['year'].isnull()))]
ratings.describe()
```

```python
ratings.corr()
```

Pearson correlation between time_diff and rating seems to be practically not existing. Let's analyse it as follows in bins - as 0-2 - 'new film', 2-12 'standard', >12 - classic. Would this influence ratings?

```python
ratings['time_diff_cat']=pd.cut(ratings['time_diff'],[0,2,12,121],right=True,labels=['new','standard','classic'], include_lowest=True)
ratings.groupby('time_diff_cat').rating.agg(['mean','count',lambda u: np.std(u,ddof=0),'median','min','max'])
```

Classic movies seems to be rated slightly higher but I'd say that the difference is negligible and there is no difference for rating in terms of production time vs rating time
